```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://www.snapdeal.com/products/electronics-headphones?sort=plrty"
```

```python
response=requests.get(url)
```

```python
response
```

```python
soup=BeautifulSoup(response.text)
```

```python
soup
```

```python
name=soup.select(".product-title")
```

```python
name
```

```python
price=soup.select(".product-price")
```

```python
price
```

```python
rating=soup.select(".product-rating-count")
```

```python
rating
```

```python
offer=soup.select(".product-discount span")
```

```python
offer
```

```python
import pandas
```

```python
headphonelist=pandas.DataFrame(name,columns=['headphone name'])
```

```python
headphonelist
```

```python
headphonelist["price"]=price
headphonelist
```

```python
headphonelist["rating"]=rating
headphonelist
```

```python
headphonelist["offer"]=offer
headphonelist
```
